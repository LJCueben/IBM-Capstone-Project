## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master
"""
conv_comp = pd.read_sql_query(QUERY,conn)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(conv_comp, bins= 10)
plt.title('Converted Compensation Histogram')
plt.xlabel('USD')
plt.ylabel('')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
Age = pd.read_sql_query(QUERY,conn)
```

```python
Age.plot(kind='box', figsize=(8, 6))

plt.title('Age Distribution')
plt.ylabel('Ages')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
age_wwhrs = pd.read_sql_query(QUERY,conn)
age_wwhrs.head()
```

```python
age_wwhrs.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='teal')

plt.title('Work Week Hours by Age Relation')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
bubble_df = pd.read_sql_query(QUERY,conn)
bubble_df.head()
```

```python
bubble_df.plot(kind='scatter',
                    x='WorkWeekHrs',
                    y='CodeRevHrs',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s= 'Age', # pass in weights 
                    xlim= (0,100)
                    )
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC LIMIT 5
"""

pie_df = pd.read_sql_query(QUERY,conn)
pie_df.set_index('DatabaseDesireNextYear', inplace=True)

pie_df
```

```python
colors_list = ['#FFB5E8','#B28DFF','#AFF8DB','#85E3FF','#E7FFAC']

pie_df['Count'].plot(kind='pie',
                            figsize=(9, 6),
                            autopct='%1.1f%%',
                            startangle=90,     
                            shadow=True,
                            colors = colors_list
                            )

plt.title('Top 5 Desired Databases')
plt.axis('equal')
plt.legend(labels=pie_df.index, loc='lower right') 


plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

stack_df = pd.read_sql_query(QUERY,conn)

stack_df['Age'] = stack_df['Age'].astype(int)

stack_df.set_index('Age', inplace=True)

median_df = stack_df.groupby('Age')[['CodeRevHrs', 'WorkWeekHrs']].median()

median_df
```

```python
median_df.plot(kind='bar',
          figsize=(8, 6), 
          color=['#BFD641','#7DE3E7'],
          stacked=True,
          )

plt.title('Median Work and Code Hours by Age')
plt.ylabel('Hours')
plt.xticks(rotation=0)
y_ticks = range(0, 46, 5) 
plt.yticks(y_ticks)

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

line_df = pd.read_sql_query(QUERY,conn)

line_df['Age'] = line_df['Age'].astype(int)

line_df.set_index('Age', inplace=True)

median_line = line_df.groupby('Age')[['ConvertedComp']].median()

median_line
```

```python
median_line.plot(kind='line',
          figsize=(8, 6), 
          color=['#C90505'],
          )

plt.title('Median Converted Compensation by Age 45-60')
plt.ylabel('Compensation in USD')
x_ticks = range(45, 61, 1) 
plt.xticks(x_ticks)

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*)
FROM master
GROUP BY MainBranch
"""

bar_df = pd.read_sql_query(QUERY,conn)

bar_df.set_index('MainBranch', inplace=True)

bar_df
```

```python
bar_df.plot(kind='barh', 
            figsize=(10, 6),
            legend=False,
            color=['#252195']
           ) 

plt.show()
```
